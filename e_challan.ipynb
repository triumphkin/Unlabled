```python
import pandas as pd
```

```python
df=pd.read_csv('baseline_chaos_data.csv')
challan_list=pd.read_csv('Challan_list.csv')
```

```python
challan_df=df.loc[df['Lane']==0,['Car_ID']].copy()
new_df=challan_df.drop_duplicates(subset=['Car_ID']).copy()

new_df=new_df.reset_index(drop=True)

new_df.head()
```

```python


# --- Sample Data (To visualize the process) ---

for target_carid in new_df['Car_ID']:
  

# --- The Core Logic ---
    
    # 2. Check if the carid already exists in challan_df to update it
    if target_carid in challan_list['Car_ID'].values:
        # Use .loc[row_condition, column_name] to safely increment the value by 1
        challan_list.loc[challan_list['Car_ID'] == target_carid, 'Challan'] += 1
        
    # 3. (Optional but recommended) What if it's not in challan_df yet?
    else:
        # If it's a first-time offender, add them to the challan_df with 1 challan
        new_offender = pd.DataFrame({'Car_ID': [target_carid], 'Challan': [1]})
        challan_list = pd.concat([challan_list, new_offender], ignore_index=True)

challan_list.to_csv('Challan_list.csv', index=False)
```
